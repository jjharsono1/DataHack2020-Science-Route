# DATA CLEANING

```python
# dropping columns relating to head of families 
population = pd.read_csv("SD1970_population.csv")
population = population.drop([i for i in population.columns if "Total persons " in i], axis = 1)
population = population.drop([i for i in population.columns if "Black persons" in i], axis = 1)
population = population.drop([i for i in population.columns if "Persons under 18" in i], axis = 1)
population = population.drop([i for i in population.columns if "Persons 65" in i], axis = 1)
population = population.replace('...', np.NaN)
population = population.dropna(subset = ['Place Name'])
```

For the next few cells, we are aggregating Male and Female ages so that we can keep age bins consistent

```python
# Aggregate Male ages to keep it Consistent
for column in population.iloc[:,10:18]:
    population = population.dropna(subset = [column])
    population[column] = population[column].astype(str).astype(int)
population["Male persons age 15-24 years"] = population.iloc[:,10:18].sum(axis = 1)
population = population.drop(population.iloc[:,10:18].columns, axis = 1)

```

```python
# Aggregate Male ages to keep it Consistent
for column in population.iloc[:,5:10]:
    population = population.dropna(subset = [column])
    population[column] = population[column].astype(str).astype(int)
population["Male persons age 5-14 years"] = population.iloc[:,5:10].sum(axis = 1)
population = population.drop(population.iloc[:,5:10].columns, axis = 1)
```

```python
# Aggregate Female ages to keep it Consistent
for column in population.iloc[:,14:19]:
    population = population.dropna(subset = [column])
    population[column] = population[column].astype(str).astype(int)
population["Female persons age 5-14 years"] = population.iloc[:,14:19].sum(axis = 1)
population = population.drop(population.iloc[:,14:19].columns, axis = 1)
```

```python
# Aggregate Female ages to keep it Consistent
for column in population.iloc[:,14:22]:
    population = population.dropna(subset = [column])
    population[column] = population[column].astype(str).astype(int)
population["Female persons age 15-24 years"] = population.iloc[:,14:22].sum(axis = 1)
population = population.drop(population.iloc[:,14:22].columns, axis = 1)
```

Here, we are reordering columns to group bins in order

```python
# Reordering columns
cols = list(population)
cols.insert(5, cols.pop(cols.index("Male persons age 15-24 years")))
cols.insert(5, cols.pop(cols.index("Male persons age 5-14 years")))
cols.insert(16, cols.pop(cols.index("Female persons age 5-14 years")))
cols.insert(17, cols.pop(cols.index("Female persons age 15-24 years")))
population = population.loc[:, cols]
```

Creating Total Person and reording those columns

```python
# Creating Total Persons
population['Total Male Persons'] =  (population[[i for i in population.columns if "Male persons age" in i]]
               .astype(str).astype(int).sum(axis = 1))

population['Total Female Persons'] =  (population[[i for i in population.columns if "Female persons age" in i]]
               .astype(str).astype(int).sum(axis = 1))
```

```python
# Reordering columns
cols = list(population)
cols.insert(4, cols.pop(cols.index("Total Male Persons")))
cols.insert(5, cols.pop(cols.index("Total Female Persons")))
population = population.loc[:, cols]
```

Dropping Columns that we don't want to work with

```python
population = population.drop(population.iloc[:,28:60].columns, axis = 1)
population = population.drop(population.iloc[:,42:].columns, axis = 1)
```

Changing all types to integers so that we can graph

```python
for column in population.iloc[:, 28:]:
    population[column] = list(map(lambda x: x.replace(',',''),list(population[column])))
for column in population.iloc[:, 4:]:
    population[column] = population[column].astype(int)
```
