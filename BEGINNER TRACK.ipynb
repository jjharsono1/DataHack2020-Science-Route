# DATA CLEANING

```python
# dropping columns relating to head of families 
population = pd.read_csv("SD1970_population.csv")
population = population.drop([i for i in population.columns if "Total persons " in i], axis = 1)
population = population.drop([i for i in population.columns if "Black persons" in i], axis = 1)
population = population.drop([i for i in population.columns if "Persons under 18" in i], axis = 1)
population = population.drop([i for i in population.columns if "Persons 65" in i], axis = 1)
population = population.replace('...', np.NaN)
population = population.dropna(subset = ['Place Name'])
population
```

```python
# Aggregate Male ages to keep it Consistent
for column in population.iloc[:,9:18]:
    population = population.dropna(subset = [column])
    population[column] = population[column].astype(str).astype(int)
population["Male Ages 15 to 24"] = population.iloc[:,9:18].sum(axis = 1)
population = population.drop(population.iloc[:,9:18].columns, axis = 1)

```

```python
male_filter = [i.split(" ")[0] == 'Male' and 'age' in i for i in population.columns]
```
